```python
# Zufllige Generierung eines kantengewichteten ungerichteten Graphen

import networkx as nx
import matplotlib.pyplot as plt
import random

def generate_random_weighted_graph(num_nodes=6, num_edges=10, min_weight=1, max_weight=20, seed=None):
    if seed is not None:
        random.seed(seed)
    G = nx.Graph()
    G.add_nodes_from(range(num_nodes))
    possible_edges = [(i, j) for i in range(num_nodes) for j in range(i+1, num_nodes)]
    random.shuffle(possible_edges)
    for (u, v) in possible_edges[:num_edges]:
        weight = random.randint(min_weight, max_weight)
        G.add_edge(u, v, weight=weight)
    return G

# Beispiel: Erzeuge einen zuflligen Graphen
G = generate_random_weighted_graph(num_nodes=7, num_edges=12, min_weight=1, max_weight=10, seed=42)

# Zeichnen
pos = nx.spring_layout(G, seed=42)
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=600)
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
plt.title("Zuflliger kantengewichteter Graph")
plt.show()
```
